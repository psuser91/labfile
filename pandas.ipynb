```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [2,4,1]
plt.plot(x,y)
plt.xlabel("Number in horizontal line ",fontsize=20)
plt.ylabel("Number in vertical line ", fontsize=20)
plt.title("line graph ", fontsize=23)
plt.show()
```
